```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("input",header=None, names=['data'])
#print(data)
```

```python
data["data"] = data["data"].str.replace("\D","", regex=True)
print(data)
```

```python
summe = 0
for i,entry in enumerate(data["data"]):
    value_str = str(entry)
    if len(value_str) == 1:
        value = int(value_str) * 11
    else:
        value_str = value_str[0], value_str[len(value_str)-1]
        #print(i,value_str, end=" ")
        value = 10 * int(value_str[0]) + int(value_str[1])
        #print(value)
    summe += value

print(summe)
```
